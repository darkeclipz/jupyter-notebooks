# Gradient descent algorithm

Gradient descent is based on the observation that if the multi-variable function $F(x)$ is defined and differentiable in a neighborhood of point $a$, then $F(x)$ decreases fastest if one goes from $a$ in the direction of the negative gradient of $F$ at $a$, $-\nabla F(a)$.

```python
%pylab inline
```

```python
import numpy as np
```

## The function $f$

Let's apply the gradient descent algorithm to find a local minimum of the function $f(x)=x^4-3x^3+2$.

```python
def f(x):
    return x**4-3*x**3+2
```

If we plot this function:

```python
points = np.arange(-10,12,0.5)
plot(points, [f(x) for x in points], color='b', linewidth=2)
grid(1)
```

## Derivative of $f$

The derivative of $f(x)$ is $f'(x)=4x^3-9x^2$.

```python
def df(x):
    return 4*x**3-9*x**2
```

Plotting both functions yields:

```python
plot(points, [f(x) for x in points], color='b', linewidth=2)
plot(points, [df(x) for x in points], color='lightblue', linewidth=2)
grid(1)
legend(["f(x)","f'(x)"]);
```

## Gradient descent

Implementation of the gradient descent algorithm:

```python
cur_x = 11
gamma = 0.00001
precision = 0.0001
previous_step_size = cur_x
steps_taken = []

while previous_step_size > precision:
    steps_taken.append(cur_x)
    prev_x = cur_x
    cur_x += -gamma * df(prev_x)
    previous_step_size = abs(cur_x - prev_x)
    
print("The local minimum occurs at %f and found after %i steps." % (cur_x, size(steps_taken)))
```

We can visualize what is going on by plotting $f(x)$, $df(x)$, and the descent to the minimum point:

```python
plot(points, [f(x) for x in points], color='b', linewidth=2)
plot(points, [df(x) for x in points], color='lightblue', linewidth=2)
plot(steps_taken, [f(x) for x in steps_taken], color='r', linewidth=3)
scatter(steps_taken[0], f(steps_taken[0]), color='r')
scatter(steps_taken[-1], f(steps_taken[-1]), color='salmon')
legend(["f(x)", "f'(x)", "descent", "start", "min."]);
```

Pretty neat trick, and the coolest thing is that it works in any $n$-dimensional space!
