# Queueing Theory -- M/M/1 model

```python
def MM1_rho(lambd, mu):
    return lambd / mu
```

```python
def MM1_L(lambd, mu):
    rho = MM1_rho(lambd, mu)
    return rho / (1 - rho)
```

```python
def MM1_Lq(lambd, mu):
    return lambd ** 2 / (mu - lambd) * mu
```

```python
def MM1_Pn(lambd, mu, n):
    rho = MM1_rho(lambd, mu)
    return rho ** n * (1 - rho)
```

```python
def MM1_P0(lambd, mu):
    return 1 - lambd / mu
```

```python
def MM1_W(lambd, mu):
    return 1 / (mu - lambd)
```

```python
def MM1_Wq(lambd, mu):
    return lambd / (mu - lambd) * mu
```
