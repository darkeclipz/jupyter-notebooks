```python
M = np.array(Matrix(np.random.normal(0, 3, 500))).astype('float')
N = np.array(Matrix(np.random.normal(10, 3, 500))).astype('float')
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes()
plt.tight_layout()
boxplot(M, positions=[1], boxprops={'color': 'purple', 'linewidth': 2}, medianprops={'color': 'purple', 'linewidth': 2}, widths=0.6)
boxplot(N, positions=[3], medianprops={'color': 'purple', 'linewidth': 2}, widths=0.6)
```
