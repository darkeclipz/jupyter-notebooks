```python
import numpy as np
```

```python
def f(x):
    return 2*x**2

points = np.arange(0,25,0.5)
plot(points, f(points), color='b', linewidth=2);
```

```python
def fPrime(x):
    return 4*x

plot(points, fPrime(points), color='b', linewidth=2);
```

## Linear

```python
for i in range(4):
    plot(points, f(points+i*10), linewidth=2);
```

## Quadratic

```python
def f(x):
    return 4*x**2+3
```

```python
plot(points, f(points), color='b', linewidth=2);
```

## Cubic

```python
def f(x):
    return 4*(x-12)**3-2**2+10
```

```python
plot(points, f(points), color='b', linewidth=2);
```

## $f'(x)$

```python
def fPrime(x):
    return 12*(x-12)**2-4*x
```

```python
plot(points, fPrime(points), color='b', linewidth=2);
```

## $f''(x)$

```python
def fPrime2(x):
    return 24*x-4
```

```python
plot(points, fPrime2(points), color='b', linewidth=2);
```

## $f'''(x)$

```python
def fPrime3(x):
    return x*0+24
```

```python
plot(points, fPrime3(points), color='b', linewidth=2);
```
