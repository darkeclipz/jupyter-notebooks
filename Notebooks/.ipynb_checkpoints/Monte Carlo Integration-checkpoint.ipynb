Generalized form of the Monte Carlo estimator:

$$ \int_a^b f(x) dx \approx \dfrac{1}{N} \sum_{i=0}^{N-1} \dfrac{f(X_i)}{pdf(X_i)}. $$

where $N$ is the number of samples, $X_i = a + \xi (b - a)$, and $\xi$ is a randomly distributed between zero and one. If we sample a uniform random variable, then $pdf(X_i) = 1 / (b-a)$.

```python
f = lambda x: 1 / (x**2 + 1)
```

```python
def monte_carlo(f, a, b, N):
    samples = [f(a + rand() * (b-a)) * (b-a) for _ in range(N)]
    return sum(samples) / N
```

```python
monte_carlo(f, -15, 15, 10000) - 2*math.atan(15)
```

```python
a = -15
b = 15
N = 512
XX = []
YY = []
SX = []
SS = []
SY = []
pdf = 1 / (b-a)
for _ in range(N):
    x = a + rand() * (b-a)
    y = f(x)
    XX.append(x)
    YY.append(y)
    SS.append(f(x) / pdf)
    if len(SS) % 10 == 0:
        R = (1 / len(SS) * sum(SS) - 2*math.atan(b))**2
        SX.append(len(SS))
        SY.append(R)
```

```python
X = np.linspace(a, b, 200)
Y = f(X)
plot(X,Y, c='black', lw=2)
plot(XX,YY, 'x', c='r', ms=10)
```
