The equation for a reduced Cauchy probability density function is:

$$ f(x) = \dfrac{1}{\pi(1+x^2)} $$

The function has fatter tails compared to a Gaussian distribution.

```python
%pylab inline
import math
```

```python
def cpd(x): return 1 / (math.pi * (1 + x**2))
```

```python
x = np.arange(-10, 10, 0.1)
y = cpd(x)
```

```python
plt.figure(figsize=(20,10))
plt.fill_between(x,y, alpha=0.25)
plt.grid()
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('CPD(x)')
plt.title('Reduced Cauchy Probability Density function');
```
